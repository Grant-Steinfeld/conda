## Word count with python collections library
[read the official documentation here](https://docs.python.org/3.8/library/collections.html#collections.Counter)

```python
import collections
import urllib
```

```python
url = 'http://agentidea.com/moby_dick.txt'
```

```python
corpus = urllib.request.urlopen(url).read().decode('utf8')
```

```python
words = corpus.split()
```

```python
k=5
```

```python
c = collections.Counter(words)
```

```python
c.most_common(k)
```

### Least common words

```python
c.most_common()[:-k-1:-1] 
```

### Words of length > 13

```python
[w for w in c.most_common() if len(w[0]) > 13]
```

```python
tens = [w for w in c.most_common() if len(w[0]) == 10]
```
