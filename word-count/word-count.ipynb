## Word count with python lib collections.Counter
[read the official documentation here](https://docs.python.org/3.8/library/collections.html#collections.Counter)

```python
import collections
import urllib
import os
```

```python
url = 'http://agentidea.com/moby_dick.txt'
```

```python
corpus = urllib.request.urlopen(url).read().decode('utf8')
```

```python
words = corpus.split()
```

```python
f"Moby Dick text has aproximately {len(words)} words"
```

```python
k=5
```

```python
c = collections.Counter(words)
```

```python
c.most_common(k)
```

### Least common words

```python
c.most_common()[:-k-1:-1] 
```

### Words of length > 13 top 7

```python
[w for w in c.most_common() if len(w[0]) > 13][:7]
```

```python
tens = [w for w in c.most_common() if len(w[0]) == 10][:5]
```

### Words with 10 characters top 5

```python
tens
```

```python
import matplotlib.pyplot as plt
```

### Plot top 4 words with length of 10

```python
plt.scatter(*zip(*tens[:4]))
```

### Words len 15 top 3 without dashes

```python
[w for w in c.most_common() if ((len(w[0]) == 15) and  (str(w[0]).find('-') == -1) and  (str(w[0]).find('â€”') == -1))][:8]
```

### The Book Moby Dick, written by `Herman Melville` is an amazing story about a whale, how often does the word `whale` occur?

```python
word_whale = [w for w in c.most_common() if w[0].lower() == 'whale']
```

```python
word_whale
```

```python
number_times_whale_occurs = sum([c[1] for c in word_whale])
```

```python
f"The word whale occurs {number_times_whale_occurs} times"
```
